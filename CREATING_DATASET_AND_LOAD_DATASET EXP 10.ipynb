#CREATING DATASET AND LOAD DATASET

```python
# Dataset creation
import pandas as pd

data= {
    "Roll_No": [101,102,103,104,105],
    "Gender":["Female","Male","Female","Male","Female"],
    "Department":["Computer","IT","ENTC","Mechanical","Computer"],
    "CGPA":[8.2,7.5,9.1,6.8,8.7],
    "Name":["Darsh","Aadi","Viraj","Vidhaan","Saksham"]
}
df = pd.DataFrame(data)

df.to_csv("students.csv", index=False)
df
```

```python
#Finding shape
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# CGPA OF ROLL_NO 105
cgpa_105 = df[df['Roll_No'] == 105]['CGPA'].iloc[0]
print(f"The CGPA of Roll_No 105 is: {cgpa_105}")
```

```python
# CGPA OF TOPPER
topper = df.loc[df['CGPA'].idxmax()]
print(f"The topper is {topper['Name']} with a CGPA of {topper['CGPA']}.")
```

```python
# DELETE GENDER COLUMN
df = df.drop('Gender', axis=1)
display(df)
```

```python
# INSERT A NEW COLUMN GRADE
def assign_grade(cgpa):
    if cgpa >= 9.0:
        return 'A+'
    elif cgpa >= 8.0:
        return 'A'
    elif cgpa >= 7.0:
        return 'B+'
    elif cgpa >= 6.0:
        return 'B'
    else:
        return 'C'

df['GRADE'] = df['CGPA'].apply(assign_grade)
display(df)
```

```python
# PERFORM MEAN, MEDIAN, MODE OPERATION

print("Mean of numerical columns:")
display(df.mean(numeric_only=True))

print("\nMedian of numerical columns:")
display(df.median(numeric_only=True))

print("\nMode of each column:")
for column in df.columns:
    print(f"Mode of {column}:")
    display(df[column].mode())
```

```python
# Access information of ROLL_NO 103
roll_no_103_info = df[df['Roll_No'] == 103]
display(roll_no_103_info)
```

```python
# Access information of Column Department
print("Unique departments and their counts:")
print(df['Department'].value_counts())

print("\nAll unique departments:")
print(df['Department'].unique())
```
